```python
import numpy as np 
import matplotlib.pyplot as plt 
from matplotlib.animation import ArtistAnimation
import IPython.display as display
from IPython.display import HTML

X = np.arange(2)
Y = np.repeat(0,2)

def trans(X, beta):
    X = X * np.sqrt(1-beta**2)
    return X

frames = []
fig = plt.figure()
ax = plt.subplot()
ax.set_xlim(0,1.2)
ax.set_ylim(-4,4)

for beta in np.arange(0,1.0, 0.01):
    text = ax.text(0.6, -3.5, f"β = {beta:.2f}")
    text1 = ax.text(0.8, -3.5, f"x = {X[1]:.1e}")
    frame = ax.scatter(X, Y, c="blue")
    X = trans(X,beta)
    frames.append([frame] + [text] + [text1])
    
animation = ArtistAnimation(fig, frames, interval=100)
html = HTML(animation.to_jshtml())
display.display(html)
plt.close()
```
